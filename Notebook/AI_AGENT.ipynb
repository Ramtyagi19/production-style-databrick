```python
from pyspark.sql import functions as F

# Read Gold Tables
store_sales = spark.read.table(
    "retail_catalog1.gold.store_sales_summary"
)

payment_summary = spark.read.table(
    "retail_catalog1.gold.payment_summary"
)

daily_sales = spark.read.table(
    "retail_catalog1.gold.daily_sales_summary"
)

print("===== TOP PERFORMING STORE =====")

top_store = (
    store_sales
    .orderBy(F.desc("total_sales"))
    .limit(1)
)

display(top_store)

print("===== PAYMENT METHOD ANALYSIS =====")
display(payment_summary)

print("===== DAILY SALES TREND =====")
display(daily_sales)

print("===== BUSINESS INSIGHTS =====")

total_sales = store_sales.agg(
    F.sum("total_sales")
).collect()[0][0]

print(f"Total Revenue: ₹{total_sales:,.2f}")

best_store = top_store.collect()[0]["store_id"]

print(f"Best Performing Store: {best_store}")

print("AI Agent Execution Completed Successfully")
```
